```python
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$unwind" : "$countries" },
                              {"$match" : {'countries': 'USA'}},
                              {"$group" : {"_id": {"cast": "$cast","countries": "$countries"}}}
                              ])
df = pd.DataFrame(list(result))
df
```

Gli attori che hanno recitato in film italiani e francesi

```python
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$unwind" : "$countries" },
                              {"$match" :{ "countries" : { "$in": ['Italy', 'USA'] }}},
                              {"$group" : {"_id": {"cast": "$cast","countries": "$countries"}}}
                              ])
df = pd.DataFrame(list(result))
df
```

I film in cui hanno recitato Margaret Singer e Sarah Ellen Stephens

```python
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$match" :{ "cast" : { "$in": ['Margaret Singer', 'Sarah Ellen Stephens'] }}},
                              {"$group" : {"_id": {"cast": "$cast","title": "$title"}}}
                              ])
df = pd.DataFrame(list(result))
df
```

7. Il numero di film per ogni attore. Ordinare in ordine decrescente a



```python
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$group" : {"_id":"$cast",  "count": { "$sum": 1 }}},
                              { "$sort": { "count": -1 }}
                              ])
df = pd.DataFrame(list(result))
df
```

8. I primi 10 attori della lista precedente

```python
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$group" : {"_id":"$cast",  "count": { "$sum": 1 }}},
                              { "$sort": { "count": -1 }},
                              { "$limit" : 10 }
                              ])
df = pd.DataFrame(list(result))
df
```

9. Gli attori che hanno recitato in un numero di film compreso tra 1° e 20

```python
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$group" : {"_id":"$cast",  "count": { "$sum": 1 }}},
                              {"$match": {"count" : {"$gt" : 0}, "count": { "$lt": 21 }} },
                              { "$sort": { "count": -1 }}
                              ])
df = pd.DataFrame(list(result))
df
```

11. Il numero medio di film in cui ha recitato un attore



```python
 
result = db.movies.aggregate([
                              {"$unwind" : "$cast" },
                              {"$group" : {"_id":"$cast",  "count": { "$sum": 1 }}},
                              {"$group": {"_id": "null", "media": {"$avg": "$count"}}},
                              ])
df = pd.DataFrame(list(result))
df
```

12. Il film con più attori

```python
result = db.movies.aggregate( [
  { "$unwind" : "$cast" },
  { "$group" : { "_id" : {"_id ": "$_id", "title" : "$title"}, "numAttori" : { "$sum" : 1 }}},
  { "$sort" : { "numAttori" : -1 } }
] )
df = pd.DataFrame(list(result))
df[df.numAttori == df.numAttori.max()]
```
